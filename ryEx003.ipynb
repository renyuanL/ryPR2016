```python
import numpy as np
import re

aDic= {'Iris-setosa':    0, 
       'Iris-versicolor':1, 
       'Iris-virginica': 2}

def getData(X):
    xL= re.split('\n',X)
    yL= [x.split(',') for x in xL]

    while [''] in yL:
        yL.remove([''])
    len(yL), yL

    zLL= []
    L=  []
    for y in yL:
        zL= []
        for i in range(len(y)-1):
            z= float(y[i])
            zL += [z]
        zLL += [zL]
        label= aDic[y[-1]]
        L  += [label]
    xA= np.array(zLL)  # data
    L=  np.array(L)    # label
    return xA, L

Y, L= getData(X)
Y, L
```

```python
# 進階 3d 畫圖

from mpl_toolkits.mplot3d import Axes3D
#from matplotlib import cm

import numpy as np
import pylab as pl


f= pl.figure()

a1= f.add_subplot(221, projection='3d')
a2= f.add_subplot(222, projection='3d')
a3= f.add_subplot(223, projection='3d')
a4= f.add_subplot(224, projection='3d')


# 以下 把 球表面座標點 算出。

#
# 球表面座標點
# 如何計算？？
#
u= np.linspace(-np.pi/2,   np.pi/2, 21) # 高度角
v= np.linspace(0, 2*np.pi, 21)          # 方位角

x= np.outer(np.cos(u),  np.cos(v))
y= np.outer(np.cos(u),  np.sin(v))
z= np.outer(np.sin(u),  np.ones(np.size(v)))

# 4 種 3d 畫圖方式

a1.scatter(x, y, z)
a2.contour(x, y, z) #, cmap=cm.coolwarm)
a3.plot_surface(x, y, z) #, rstride=4, cstride=4, color='b')
a4.plot_wireframe(x, y, z)

pl.show()
```
